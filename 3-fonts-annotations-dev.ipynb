```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

tips = sns.load_dataset('tips')
tips.head()
```

```python
for context in ['paper', 'notebook', 'talk', 'poster']:
    sns.set_context(context)
    sns.displot(data=tips, x='tip')
    plt.show()
```

```python
# Reset styles
sns.reset_orig()
```

```python
sns.set_context("talk", font_scale=1.1, rc={"grid.linewidth": 10})
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
# Reset styles
sns.reset_orig()
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)
plt.xlabel('Day of week')
plt.ylabel('Total bill (excludes tip)')
```

```python
avg_tip = np.mean(tips['tip'])

sns.displot(x='tip', data=tips)
plt.axvline(np.mean(tips['tip']), lw=4, c='gray')
plt.title('Average tip: $' + str(np.round(avg_tip,3)))
```

```python
tip_corr = stats.pearsonr(tips['tip'], tips['total_bill'])
tip_corr
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.annotate(text='Correlation coefficient: ' + str(np.round(tip_corr[0],2)), xy=(10, 8))
```
