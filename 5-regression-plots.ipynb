```python
import seaborn as sns

penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
penguins['sex01'] = penguins['sex'].replace(['Female','Male'],[0,1])
penguins['sex01'].unique()
```

```python
sns.regplot(x='bill_length_mm', y='sex01', data=penguins)
```

```python
sns.regplot(x='bill_length_mm', y='sex01', data=penguins, logistic=True)
```

```python
sns.regplot(x='bill_length_mm', y='body_mass_g', data=penguins, lowess=True)
```

```python
sns.lmplot(x='bill_length_mm', y='body_mass_g', data=penguins, hue='species')
```

```python
sns.lmplot(x='bill_length_mm', y='body_mass_g', data=penguins, col='species')
```

```python
sns.lmplot(x='bill_length_mm', y='body_mass_g', data=penguins, col='species', hue='sex')
```

```python
sns.lmplot(x='bill_length_mm', y='sex01', data=penguins, col='species', logistic=True)
```
