```python
import seaborn as sns
import matplotlib.pyplot as plt

mpg = sns.load_dataset('mpg')
mpg.head()
```

```python
sns.swarmplot(x='origin', y='acceleration', data=mpg)
```

```python
sns.set_style('dark')
sns.swarmplot(x='origin', y='acceleration', data=mpg)
```

```python
sns.set_style('white')
sns.swarmplot(x='origin', y='acceleration', data=mpg)
```

```python
sns.reset_orig()
```

```python
sns.swarmplot(x='origin', y='acceleration', data=mpg)
sns.despine()
```

```python
# These come from matplotlib
sns.boxplot(x='origin', y='mpg', color='b', data=mpg)
```

```python
sns.set_palette('pastel')
sns.boxplot(x='origin', y='mpg', data=mpg)
```

```python
sns.set_palette('Blues')
sns.boxplot(x='origin', y='mpg', data=mpg)
```
