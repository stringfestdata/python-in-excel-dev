```python
import seaborn as sns

mpg = sns.load_dataset('mpg')
mpg.head()
```

```python
# Violin plot
sns.violinplot(data=mpg, x='horsepower')
```

```python
sns.violinplot(data=mpg, x='origin', y='horsepower')
```

```python
# Strip plot 
sns.stripplot(data=mpg, x='mpg')
```

```python
sns.stripplot(data=mpg, x='mpg', y='origin')
```

```python
sns.stripplot(data=mpg, x='mpg', y='origin', jitter=.3)
```

```python
sns.swarmplot(data=mpg, x='origin', y='mpg')
```

```python
sns.swarmplot(data=mpg, x='origin', y='mpg', hue='cylinders')
```
