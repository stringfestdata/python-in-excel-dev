```python
import seaborn as sns

mpg = sns.load_dataset('mpg')
mpg.head()
```

```python
g = sns.FacetGrid(mpg, col='origin')
g.map(sns.boxplot, 'mpg')
```

```python
g = sns.FacetGrid(mpg, col='origin',  row='cylinders')
g.map(sns.scatterplot, 'weight', 'horsepower')
```

```python
sns.catplot(x='origin', y='mpg', data=mpg)
```

```python
sns.catplot(x='origin', y='mpg', data=mpg, kind='violin')
```

```python
sns.catplot(x='origin', y='mpg', data=mpg, kind='violin', col='model_year')
```

```python
sns.catplot(x='origin', y='mpg', data=mpg, kind='violin', row='model_year')
```

```python
sns.relplot(x='weight', y='mpg', data=mpg, col='origin')
```

```python
sns.relplot(x='model_year', y='mpg', data=mpg, kind='line', col='origin')
```
